# Problemas de Estadística Descriptiva

## Ejercicio 1 – ¿Qué tan largos son los títulos de las películas?

En este ejercicio vas a aplicar técnicas de estadística descriptiva sobre un conjunto de datos real: una base con 1000 películas populares del sitio IMDb.

A partir del título de cada película, vas a:
- Calcular la longitud del título (en caracteres).
- Obtener métricas estadísticas clave: media, mediana, moda, rango, varianza, desviación estándar, asimetría y curtosis.
- Identificar los títulos más largos y más cortos.
- Visualizar la distribución de estas longitudes con un histograma y una curva de densidad.


```python
# Importar librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis, mode
import seaborn as sns

# Cargar dataset de IMDb
url = "../assets/imdb_1000.csv"
df = pd.read_csv(url)
df
```

```python
# Crea una nueva columna llamada 'title_length' con la longitud (número de caracteres) de cada título
df['title_length']=df['title'].str.len()
df
```

```python
# Calcula media, mediana y moda de la columna 'title_length'
mean=df['title_length'].mean()
median=df['title_length'].median()
mode=df['title_length'].mode().iloc[0]
print(f'Media:{mean} Mediana: {median} Moda: {mode}')
```

```python
# Calcula el rango, la varianza y la desviación estándar de 'title_length'
range_ = df['title_length'].max() - df['title_length'].min()
variance = df['title_length'].var()
std = df['title_length'].std()
cv=(std/mean)*100
print(f'Rango: {range_} Varianza: {variance} Desviacion estandar: {std} Coeficiente de variacion: {cv:.2f} %')
```

```python
# Calcula asimetría y curtosis de la columna 'title_length'
skewness = df['title_length'].skew()
kurt = df['title_length'].kurt()
print(f'Asimetria: {skewness} Curtosis: {kurt}')
```

```python
# Imprime el título más corto y el título más largo según su longitud
tituloMasCorto=df[df['title_length'] == df['title_length'].min()]['title'].iloc[0]
tituloMasLargo=df[df['title_length'] == df['title_length'].max()]['title'].iloc[0]
print(f'Titulo mas largo: {tituloMasLargo} , Titulo mas corto: {tituloMasCorto} ')
```

```python
# grafica tus resultados
df['title_length'].hist(bins=40)
plt.title('Histograma de la Distribución de title_length')
plt.xlabel('title_length')
plt.ylabel('Frecuencia')
plt.axvline(mean, color='red', linestyle='--', label=f'Media: {mean:.2f}')
plt.axvline(median, color='green', linestyle='--', label=f'Mediana: {median}')
plt.axvline(mode, color='orange', linestyle='--', label=f'Moda: {mode}')
plt.legend()
plt.show()
```

De acuerdo a los datos obtenidos tenemos que una asimetria positiva por ende los datos estan mayormente agrupados hacia la izquierda, asi mismo presenta uns distribucion leptocúrtica, es decir, tiene un pico más agudo y colas más pesadas que la normal.
tambien al analizar los coeficiente de variacion podemos observar que los datos se alejan bastante de la media.

## Ejercicio 2 – Cálculo manual de desviación estándar

En este ejercicio vas a calcular manualmente la desviación estándar, seleccionando cinco títulos reales del dataset de películas que cargaste antes.


```python
# Crea una lista con 5 valores reales de df['title_length'], por ejemplo: [10, 13, 14, 18, 22]
dfnueva=df.iloc[[10, 13, 14, 18, 22]][['title','title_length']]


dfnueva
```

```python
# Calcula la media de los 5 valores
mean=sum(dfnueva['title_length'])/dfnueva['title_length'].count()
print(mean)
```

```python
# Calcula las diferencias al cuadrado con respecto a la media
diferencia_al_cuadrado=(dfnueva['title_length']-mean)**2
diferencia_al_cuadrado
```

```python
# Calcula la varianza (promedio de las diferencias al cuadrado) y su raíz cuadrada para obtener la desviación
variance=sum(diferencia_al_cuadrado)/dfnueva['title_length'].count()
std=np.sqrt(variance)
cv=(std/mean)*100

print (f'varianza: {variance:.2f} desviacion estandar: {std:.2f}  coeficiente de variacion: {cv:.2f} %')
```
